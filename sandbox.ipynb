```python
import pandas as pd
import numpy as np
from bokeh.charts import Histogram, Bar, output_notebook, show
output_notebook()

allmedia = pd.read_csv('mediasources.csv')
allmedia
```

```python
# national-level sources
print [i for i in sorted(set(allmedia['region'])) if i.startswith('Nat')] # catch misspellings
nationals = ['Natiional', 'Natioanl', 'National', 'Nationalp', 'Nationsl', 'Nationwide', 'Natonal']
nationalsources = [s for s in allmedia.index.values if allmedia['region'][s] in nationals]
allmedia.loc[nationalsources]
```

```python
# every country, individually
countrydict = {}
for country in sorted(set(allmedia['country'])):
    countrydict[country] = allmedia[allmedia['country'] == country]
```

```python
counts = [len(countrydict[i]) for i in sorted(countrydict.keys())] # sources per country
natcounts = [] # national sources per country
for i in sorted(countrydict.keys()):
    get = [n for n in nationalsources if n in countrydict[i].index]
    natcounts += [len(countrydict[i].loc[get])]
sourcecounts = pd.DataFrame({'country':sorted(countrydict.keys()),
                             'numsources':counts,
                             'numnationalsources': natcounts})
sourcecounts.set_index('country', inplace = True)
sourcecounts.to_csv('mediasource_counts.csv')
print sourcecounts.describe()
```

```python
p = Histogram(sourcecounts['numsources'], title="number of sources per country")
show(p)
```
